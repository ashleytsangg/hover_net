```python
import seaborn as sn
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
```

```python
mat_path = '../FTE302-20X-STIC_class_acc.mat'
mat = scipy.io.loadmat(mat_path)
basename = 'FTE302-20X-STIC_class_acc'
acc = round(mat['acc'][0][0] * 100, 2)
```

```python
from sklearn.metrics import confusion_matrix

# print(labeled_ind)
y_true = mat['truth']
y_pred = mat['pred']

cm = confusion_matrix(y_true, y_pred)
```

```python
g = sn.heatmap(cm, annot=True, fmt='.5g')
plt.xlabel('Pred Class')
plt.ylabel('True Class')
g.set_xticklabels(np.arange(0,6))
plt.title('FTE302-20X-STIC, acc ={}%'.format(acc))
fn = basename + '-50.png'
plt.savefig(fn)
```
